```python
#Générer de façon aléatoire une carte de crédit(numéro, date d'expiration, code de sécurité)
fake = Faker(locale = "fr_FR")
for _ in range(10):
    print(f"Numéro de la carte : {fake.credit_card_number()},Date d'expiration : {fake.credit_card_expire()},Code de sécurité : {fake.credit_card_security_code()}")
```

```python
# Générer les couleurs de façon aléatoire
fake = Faker(locale = "fr_FR")
for _ in range(10):
    print(f" Rouge, vert, bleu : {fake.rgb_color()}")
    print(f"hex_color : {fake.hex_color()}")
```

```python
# Générer les ISBN 
fake = Faker(locale = "fr_FR")
# remplace les nombre de 0 à 9
# % remplace les nombre de 1 à 9
for _ in range(5):
    print(fake.numerify(text = "%%%-#-%%%%-%%%%-%%%-##"))
```
